```python
# Dropping columns with null values and non-numeric values
df_train = df_train.drop(['binnedInc', 'Geography', 'PctSomeCol18_24', 'PctEmployed16_Over', 'PctPrivateCoverageAlone'], axis=1)
```

# Preparing the Selected Features

```python
# Checking for null values and replacing it with the median
df_train.dropna(inplace=True)
df_train.fillna(df_train.mean(), inplace=True)
```

```python
# Checking again for null values
df_train.isna().sum()
```

```python
# Preparing y
y = df_train.pop('TARGET_deathRate')
```

```python
# Preparing X
X = df_train
```

### 3. Assess Baseline

```python
# Calculate the average of the target variable after importing the numpy package
y_mean = y.mean()
```

```python
# Create a numpy array called y_base filled with this value
y_base = np.full(y.shape, y_mean)
```

```python
# Import the MSE and MAE from sklearn
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
```

```python
# Calculate the MSE and MAE between the target variable and the average value
print(mse(y, y_base, squared=True))
print(mae(y, y_base))
```

### 4. Train Initial Model

```python
# Import the linear regression module from sklearn
from sklearn.linear_model import LinearRegression
```

```python
# Instantiate the LinearRegression class into a variable called reg1
reg1 = LinearRegression()
```

```python
#  Fit the model with the following features: "Year", "Population", "Life_expectancy", "CO2"
reg1.fit(X, y)
```

```python
# Calculate the MSE and MAE between the target variable and the predictions
y_preds = reg1.predict(X)
print(mse(y, y_preds, squared=True))
print(mae(y, y_preds))
```

```python
#  Create a line chart with y as both x and y axes
line_chart1 = alt.Chart(pd.DataFrame({'target': y, 'preds': y})).mark_line(color='green').encode(
    x='target',
    y='preds'
)
line_chart1
```

```python
# Create a line chart with y as the x-axis and y_preds as y-axis and combine it with previous chart
line_chart2 = alt.Chart(pd.DataFrame({'target': y, 'preds': y_preds})).mark_line(color='blue').encode(
    x='target',
    y='preds'
)
line_chart2
```

```python
line_chart2 + line_chart1
```
